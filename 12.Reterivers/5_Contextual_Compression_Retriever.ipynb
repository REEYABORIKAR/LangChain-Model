```python
from langchain_community.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain.retrievers.contextual_compression import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain_core.documents import Document
```

```python
## Recreate the documet object from the previous data
docs=[
    Document(page_content=(""" The Grand Canal is one of the most visited natural builders in the world.
    Photosynthesis is the process by which green plants sculpture sunlight into energy.
    Millions of tourists travel to see it every year the rock state back millions of years.
    """),metadata={"source":"Doc1"}),
    Document(page_content=("""
    In medieval Europe, castles were built primarily for defense.
    The chlorophyll in the palace captures sunlight during photosynthesis.
    Knighs wore armor made of metal steal. Seige weapons were often used to breach castle wars"""),metadata={"source":"Doc2"}),
    Document(page_content=(""" Basketball was invented by Dr James Nigh Smith in the late 19th century.
    It was originally played with a soccer ball and Peach basket envy is now a global league"""),metadata={"source":"Doc3"}),
    Document(page_content=("""
    The history of cinema begins in the late 1880s. Silent films were the earliest form.
    Thomas Edison was among the pioneers. Photosynthesis does not occur in animal cells
    Modern filmmaking involves complex CGI and sound design"""),metadata={"source":"Doc1"})

]
```

```python
## Create FAISS vector store from the documents
embedding_model=GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
vectorstore=FAISS.from_documents(docs, embedding_model)
```

```python
base_retriever=vectorstore.as_retriever(search_kwargs={"k":5})
```

```python
## Set up the contextual compressor using LLM
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0)
compressor=LLMChainExtractor.from_llm(llm)
```

```python
## Set up the contextual compression retriever
compression_retriever=ContextualCompressionRetriever(
    base_retriever=base_retriever,
    base_compressor=compressor
)
```

```python
## Query the retriever
query="What is Photosynthesis?"
compressed_result=compression_retriever.invoke(query)
```

```python
for i, doc in enumerate(compressed_result):
    print(f"\n---Result {i+1} ---")
    print(f"Content :\n {doc.page_content}...")
```
