```python
from langchain_community.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
```

```python
## Sample documents
docs=[
    Document(page_content=" The langchain makes it easy to work with LLMs"),
    Document(page_content=" Langchain is used to build LLM based applications"),
    Document(page_content="Chroma is used to store ans search and search document embeddings"),
    Document(page_content="Embeddings are vector representationa of text"),
    Document(page_content="MMR helps you get diverse results when doing similarity search"),
    Document(page_content="Langchain supports Chroma, FAISS,Pinecone, and more"),

]
```

```python
from langchain_community.vectorstores import FAISS

## Initialize Google Embedding
embedding_model=GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

# Step 2: Create FAISS vectore store from documents
vectorstore=FAISS.from_documents(
    documents=docs,
    embedding=embedding_model
)
```

```python
## Enable MMR in the retriever
retriever=vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k":3, "lambda_mult":0.5}
)
```

```python
query="what is langchain?"
results=retriever.invoke(query)
```

```python
for i, doc in enumerate(results):
    print(f"\n---Result {i+1} ---")
    print(f"Content :\n {doc.page_content}...")
```
