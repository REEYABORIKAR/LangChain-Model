# Currency Conversation Tool

```python
## tool create
from langchain_core.tools import InjectedToolArg
from typing import Annotated

@tool
def get_conversion_factor(base_curr:str, target_curr:str)-> float:
    """This function fetches the currency factor between a given base currency and a target currency"""
    url=f'https://v6.exchangerate-api.com/v6/76c32eb9f67bda12e46cea90/pair/{base_curr}/{target_curr}'

    response=requests.get(url)
    return response.json()


@tool
def convert(base_curr_val:int, conversion_rate:Annotated[float, InjectedToolArg])->float:
    """Given a currency conversion rate this function calculates the target currency value from given base currency value"""
    return base_curr_val*conversion_rate
```

```python
convert.args
```

```python
get_conversion_factor.invoke({'base_curr':'USD','target_curr':'INR'})
```

```python
convert.invoke({'base_curr_val':100, 'conversion_rate':85.16})
```

```python
## Tool binding
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
llm_with_tools=llm.bind_tools([get_conversion_factor, convert])
```

```python
messages=[HumanMessage('What is the conversion factor between USD and INR, and based on that can you convert 10 USD to INR')]
```

```python
messages
```

```python
ai_messages=llm_with_tools.invoke(messages)
```

```python
messages.append(ai_messages)
```

```python
ai_messages
```

```python
ai_messages.tool_calls
```

```python
import json

for tool_call in ai_messages.tool_calls:
    # print(tool_call)
    # execute the 1st tool and get the value of conversion rate
    if tool_call['name']=="get_conversion_factor":
        tool_message1=get_conversion_factor.invoke(tool_call)
        # print(tool_message1)
    # fetch this conversion rate
    conversion_rate=json.loads(tool_message1.content)['conversion_rate']
    # append this tool message to message list
    messages.append(tool_message1)
    # execute the 2nd tool using the conversion rate from tool 1
    if tool_call['name']=='convert':
    # fetch current argument
        tool_call=['args']['conversion_rate']=conversion_rate
        tool_message2=convert.invoke(tool_call)
        messages.append(tool_message2)
```
